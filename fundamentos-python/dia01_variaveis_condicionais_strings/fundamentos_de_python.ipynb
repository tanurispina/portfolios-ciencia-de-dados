```python
#Exercício 3: O Classificador de Notas (Lógica Condicional)
#Vamos simular o sistema de aprovação de um curso.
#Crie uma variável chamada nota_aluno e atribua um valor
#numérico a ela (por exemplo, 6.5).
#Crie uma estrutura if, elif e else com as seguintes regras:
#Se a nota for maior ou igual a 7.0: imprima "Status: Aprovado direto!"
#Se a nota for maior ou igual a 5.0 E menor que 7.0:
# imprima "Status: Em recuperação."
#Se a nota for menor que 5.0: imprima "Status: Reprovado."
#Dica: Teste trocar o valor da variável nota_aluno algumas vezes
# e rodar a célula novamente para ver se a lógica está caindo
# na condição certa.

nota_aluno=6.5

if nota_aluno > 7.0:
  print("Status: Aprovado direto!")
elif nota_aluno >=5.0 and nota_aluno <7.0:
  print("Status: Em recuperação.")
else:
  print("Status: Reprovado.")
```

```python
#Exercício 1: Padronizando relatórios no SupriCheck
#Imagine que no módulo de relatórios do SupriCheck,
#o técnico digitou o nome do equipamento de forma totalmente despadronizada,
#com espaços sobrando e letras minúsculas: "   roteador principal fibra   ".
#Crie uma variável com esse texto bagunçado.
#Use os métodos de string para: remover os espaços das pontas,
#deixar tudo em MAIÚSCULAS e substituir a palavra "FIBRA" por "OPTICA".
#Imprima o resultado final (deve ficar "ROTEADOR PRINCIPAL OPTICA").

texto_baguncado ="   roteador principal fibra   "
texto_com_strip=texto_baguncado.strip()
texto_com_upper=texto_com_strip.upper()
texto_com_replace=texto_com_upper.replace("FIBRA", "OPTICA")
texto_manipulado = texto_com_replace
print(texto_manipulado)


texto_manipulado=texto_baguncado.strip().upper().replace("FIBRA", "OPTICA")
print(texto_manipulado)
```

```python
#Exercício 2: Filtro de Balcão de Farmácia
#Vamos criar um pequeno verificador para o estoque de uma farmácia.
#Peça para o usuário digitar o nome de um medicamento usando input(). Guarde em uma variável.
#Como as pessoas digitam de formas diferentes, converta o que o usuário digitou para
# letras minúsculas usando .lower().
#Crie uma condição (if/else): se a palavra "gotas" estiver presente (in) no texto convertido,
# imprima "Medicamento líquido detectado.". Caso contrário, imprima
#"Apresentação em comprimido ou outro formato."

remedio= input('Informe o medicamento desejado: ').lower()


if "gotas" in remedio:
  print("Medicamento líquido detectado.")
else:
  print("Apresentação em comprimido ou outro formato.")
```

```python
#Exercício 3: Análise de Notas da PUC Minas
#Vamos simular o fechamento de semestre de uma matéria da sua pós-graduação.
#Crie três variáveis com notas fictícias (ex: nota1 = 7.8, nota2 = 9.5, nota3 = 6.2).
#Calcule a média das três notas e guarde em uma variável media_final.
#Use a função round() para arredondar a media_final para apenas 1 casa decimal.
#Use o print() combinado com max() e min() para exibir qual foi a maior nota e a menor nota do
#semestre, e depois imprima a média arredondada.

nota1 = 7.8
nota2 = 9.5
nota3 = 6.2

media_final = (nota1 + nota2 + nota3) / 3
media_arredondada = round(media_final, 1)

print("A maior nota do semestre foi:", max(nota1, nota2, nota3))
print("A menor nota do semestre foi:", min(nota1, nota2, nota3))
print("A média final da disciplina é:", media_arredondada)

```
